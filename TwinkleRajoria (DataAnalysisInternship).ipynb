```python
import pandas as pd
```

```python
#Loading The Dataset
df = pd.read_csv("Amazon Sale Report.csv")
```

```python
df.head()
```

```python
# Data Cleaning: Check for missing values and data types
missing_values = df.isnull().sum()
data_types = df.dtypes

missing_values, data_types
```

```python
# Drop irrelevant columns
df_cleaned = df.drop(columns=['New', 'PendingS'])
```

```python
# Inspect the relationship between missing 'currency'/'Amount' and 'Status'
missing_amount_status = df_cleaned[df_cleaned['Amount'].isnull()]['Status'].value_counts()
```

```python
# Drop rows with missing shipping details
df_cleaned = df_cleaned.dropna(subset=['ship-city', 'ship-state', 'ship-postal-code', 'ship-country'])
```

```python
# Re-check missing values
missing_values_cleaned = df_cleaned.isnull().sum()

missing_amount_status, missing_values_cleaned
```

```python
# Remove 'Cancelled' orders
df_cleaned = df_cleaned[df_cleaned['Status'] != 'Cancelled']

# Re-check missing values after removing cancelled orders
missing_values_after_removal = df_cleaned.isnull().sum()

# Proceed with EDA: Generate summary statistics and initial visualizations
summary_statistics = df_cleaned.describe(include='all')

missing_values_after_removal, summary_statistics
```
